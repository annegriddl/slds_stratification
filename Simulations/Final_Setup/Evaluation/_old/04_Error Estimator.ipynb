## Comparsion RF vs XGB

```python
data_error_concat = pd.concat([df_rf.loc[:, ['param_model', 'error_estimator_mean_stratified', 'error_estimator_mean_unstratified', 'param_transformation'] ] , 
           df_xgb.loc[:, ['param_model', 'error_estimator_mean_stratified', 'error_estimator_mean_unstratified', 'param_transformation'] ]],
             axis=0, ignore_index=True)
print(data_error_concat.shape)
transformations = data_error_concat['param_transformation'].unique()

for transformation in transformations:
    data_error= data_error_concat[data_error_concat['param_transformation'] == transformation]
    # make data long
    data_error_long = data_error.melt(id_vars=['param_model'], 
                                               value_vars=['error_estimator_mean_stratified', 'error_estimator_mean_unstratified'], 
                                               var_name='stratification', value_name='error')
    plt.title('Boxplot of Error Estimator: ' + transformation)
    sns.boxplot(x=data_error_long['param_model'], 
                y=data_error_long['error'], 
                hue=data_error_long['stratification'], 
                palette=['#26A94C', 'darkgreen'])
    plt.legend(labels=['Stratified', 'Unstratified'])
    plt.show()

```

```python
for transformation in ['log', 'identity', 'sqrt']:
    df_transfromation = df_xgb[df_xgb['param_transformation'] == transformation]
    print(df_transfromation.shape)

    df_plot = pd.DataFrame({'Stratified': df_transfromation['error_estimator_mean_stratified'].values,
                            'Unstratified': df_transfromation['error_estimator_mean_unstratified'].values},
                           index=df_transfromation['parameter_combination_string'].values)
    
    df_plot = df_plot.sort_values(by=['Stratified', 'Unstratified'], ascending=True)

    ax = df_plot.plot.bar(rot=0, color={"Stratified": '#26A94C', "Unstratified": 'darkgreen'})
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    ax.set_title(f'XGB: Mean Error Estimator with {transformation} transformation') 
    ax.legend(loc='lower right')
```
